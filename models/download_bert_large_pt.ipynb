# Download BERT-Large Portuguese para uso Offline

**Este notebook baixa o BERT-Large Portuguese e salva como output para uso offline.**

Modelo: `netoferraz/bert-large-portuguese-cased-finetuned-peticoes`

```python
import os
from transformers import AutoTokenizer, AutoModel

# Modelo a baixar (BERT-Large Portuguese)
MODEL_NAME = "netoferraz/bert-large-portuguese-cased-finetuned-peticoes"
OUTPUT_DIR = "/kaggle/working/bert-large-portuguese"

print(f"Baixando modelo: {MODEL_NAME}")
print(f"Destino: {OUTPUT_DIR}")
```

```python
# Baixar tokenizer e modelo
print("[1/3] Baixando tokenizer...")
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)

print("[2/3] Baixando modelo (BERT-Large ~1.3GB, pode demorar)...")
model = AutoModel.from_pretrained(MODEL_NAME)

print("[3/3] Salvando em disco...")
os.makedirs(OUTPUT_DIR, exist_ok=True)
tokenizer.save_pretrained(OUTPUT_DIR)
model.save_pretrained(OUTPUT_DIR)

print("Modelo salvo em:", OUTPUT_DIR)
```

```python
# Verificar arquivos salvos
print("Arquivos salvos:")
print("-" * 50)
total = 0
for f in sorted(os.listdir(OUTPUT_DIR)):
    path = os.path.join(OUTPUT_DIR, f)
    if os.path.isfile(path):
        size = os.path.getsize(path)
        total += size
        print(f"{f:40} {size/1024/1024:>8.2f} MB")
print("-" * 50)
print(f"TOTAL: {total/1024/1024:.2f} MB")
```

```python
# Testar carregamento local
print("Testando carregamento local...")
tok_test = AutoTokenizer.from_pretrained(OUTPUT_DIR)
mod_test = AutoModel.from_pretrained(OUTPUT_DIR)

# Teste rapido
text = "Mamografia sem alteracoes suspeitas."
inputs = tok_test(text, return_tensors="pt")
outputs = mod_test(**inputs)

print("Teste OK!")
print(f"   Vocab: {tok_test.vocab_size}")
print(f"   Hidden: {mod_test.config.hidden_size}")
print(f"   Layers: {mod_test.config.num_hidden_layers}")
```
