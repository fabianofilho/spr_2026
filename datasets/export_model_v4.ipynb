# Export model-v4 → Dataset próprio

**Objetivo:** Copiar os pesos do `gabrielsavio/model-v4` para o output deste notebook,
criando uma versão vinculada à sua conta no Kaggle.

## ⚙️ Configuração no Kaggle

1. **Add Input** → Datasets → pesquisar `gabrielsavio/model-v4` → adicionar
2. **Settings → Internet**: OFF (não precisa de internet)
3. **Settings → Accelerator**: None
4. **Run All** → confirmar que não houve erros
5. **Save Version** → **Save & Run All (Commit)**
6. Após salvar: vá em **Data** (aba do notebook) → **Output** → **New Dataset**
   - Dê o nome `model-v4` e clique em **Create**
   - Aguarde a criação do dataset

## 🔗 Após criar o dataset

Seus notebooks de resubmit passarão a usar:
```python
WEIGHTS_BASE = Path('/kaggle/input/model-v4/advanced_outputs_kaggle_4')
```

````python
import os
import shutil
from pathlib import Path

# ═══════════════════════════════════════════════════════════════════════════════
# PATHS
# ═══════════════════════════════════════════════════════════════════════════════
# Fonte: dataset gabrielsavio/model-v4 (adicione como input)
SOURCE = Path('/kaggle/input/model-v4/advanced_outputs_kaggle_4')

# Destino: output do notebook (vira dataset próprio)
DEST = Path('/kaggle/working/advanced_outputs_kaggle_4')

print(f'Fonte : {SOURCE}')
print(f'Destino: {DEST}')
print(f'Fonte existe: {SOURCE.exists()}')
````

````python
# ═══════════════════════════════════════════════════════════════════════════════
# VERIFICAR ESTRUTURA DA FONTE
# ═══════════════════════════════════════════════════════════════════════════════
if not SOURCE.exists():
    raise FileNotFoundError(
        f'Dataset não encontrado em {SOURCE}.\n'
        'Adicione o dataset gabrielsavio/model-v4 como input:\n'
        '  Add Input → Datasets → model-v4 (gabrielsavio)'
    )

print('Estrutura da fonte:')
total_bytes = 0
file_list = []
for root, dirs, files in os.walk(SOURCE):
    dirs.sort()
    level = Path(root).relative_to(SOURCE).parts
    indent = '  ' * len(level)
    print(f'{indent}{Path(root).name}/')
    for f in sorted(files):
        fpath = Path(root) / f
        size = fpath.stat().st_size
        total_bytes += size
        file_list.append(fpath)
        print(f'{indent}  {f}  ({size/1024/1024:.1f} MB)')

print(f'\nTotal: {len(file_list)} arquivos, {total_bytes/1024/1024:.1f} MB')
````

````python
# ═══════════════════════════════════════════════════════════════════════════════
# COPIAR TUDO PARA O OUTPUT
# ═══════════════════════════════════════════════════════════════════════════════
print(f'Copiando {SOURCE} → {DEST}...')

if DEST.exists():
    shutil.rmtree(DEST)

shutil.copytree(SOURCE, DEST)
print('Cópia concluída!')
````

````python
# ═══════════════════════════════════════════════════════════════════════════════
# VERIFICAR INTEGRIDADE DO OUTPUT
# ═══════════════════════════════════════════════════════════════════════════════
copied_files = list(DEST.rglob('*'))
copied_size  = sum(f.stat().st_size for f in copied_files if f.is_file())

print('Verificação do output:')
print(f'  Arquivos copiados: {sum(1 for f in copied_files if f.is_file())}')
print(f'  Tamanho total:     {copied_size/1024/1024:.1f} MB')

# Checar arquivos críticos
critical = [
    'all_results.pkl',
    'solo_bert_artifacts.pkl',
    'weights/bertimbau_large__cb_focal/fold_0.pt',
    'weights/bertimbau_large__cb_focal/fold_1.pt',
    'weights/bertimbau_large__cb_focal/fold_2.pt',
    'weights/bertimbau_large__cb_focal/fold_3.pt',
    'weights/bertimbau_large__cb_focal/fold_4.pt',
    'weights/bertimbau_large__cb_focal/tokenizer/tokenizer.json',
    'weights/bertimbau_large__cb_focal/model_config/config.json',
]
print('\nArquivos críticos:')
all_ok = True
for rel_path in critical:
    full = DEST / rel_path
    ok   = full.exists()
    all_ok = all_ok and ok
    status = '✅' if ok else '❌ FALTANDO'
    print(f'  {status}  {rel_path}')

if all_ok:
    print('\n✅ Todos os arquivos críticos presentes. Pronto para salvar!')
else:
    print('\n❌ Alguns arquivos estão faltando. Verifique o dataset de origem.')
````
